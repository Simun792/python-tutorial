### Datatypes and Variables

```python
# Datatypes: int, float, str, boolean, list, tupple, dictionary, set

a = 5 # int datatype
b = 8.98 # float datatype
c = "simanchal is working in a IT company" # string datatype
d = True or False # Boolean datatype
e = [1, 2, "a", "b", "c", 8.43] # list datatype
f = (1, 2, 3, "a", "a") # tupple datatype
g = {"a": "rakesh", "b": "sima", "c": ["rakesh", "sima", 1, 2]} # dictionary datatype
h = {"a", "b", "a", "rakes", "sima", "sima"} # set datatype
```

```python
print(type(a))
print(type(b))
print(type(c))
print(type(d))
print(type(e))
print(type(f))
print(type(g))
print(type(h))
```

### Loops

```python
# For Loop: break, continue, and/or condition

for i in range(1, 31):
    if i == 20:
        break

    if i == 6 or i == 7:
        continue

    if i % 2 == 0 and i % 3 == 0:
        print(f"{i} is divisble by both 2 and 3")


    if i % 2 == 0:
        print(f"{i} is divisible by 2")
    elif i % 3 == 0:
        print(f"{i} is divisible by 3")
    elif i % 5 == 0:
        print(f"{i} is divisible by 5")
    else:
        print(f"{i} is not divisible by 2,3 and 5") 
```

```python
# While
n = 0
while n <= 20:
    print(n)
    n = n + 1
```

### Functions

```python
import math
def area_of_square(width: float, height: float) -> float:
    area = width * height
    return area

def area_of_circle(radius: float) -> float:
    area = math.pi * radius ** 2
    return area
```

```python
def calculate_area_of_square():
    h = 7
    w = 5
    area = area_of_square(width=w, height=h)
    print(f"Area of square is {area}")
```

```python
calculate_area_of_square()
```

```python
def calculate_area_of_circle():
    r = 8.45
    area = area_of_circle(radius=r)
    print(f"Area of circle is {area}")
```

```python
calculate_area_of_circle()
```

### File Handling

```python
# modes in file handling: read, write, read and write, append

# Read
with open("book.txt", "r") as f:
    data = f.read()

print(data)
```

```python
new_data = "This data will be copied to book copy file\n" + data
print(new_data)
```

```python
with open("book_copy.txt", "w") as f:
    f.write(new_data)
```
